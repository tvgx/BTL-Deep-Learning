```python
# Remove unnecessary folders and files in stable_diffusion_faces
import shutil

list_remove = ['.git', '.idea', '.gitignore', 'README.md', 'LICENSE.txt', 'sample.jpg']
for item in list_remove:
    item_path = os.path.join("../data/raw/stable_diffusion_faces/stable-diffusion-face-dataset", item)
    if os.path.isdir(item_path):
        shutil.rmtree(item_path)  # Use shutil.rmtree() for non-empty directories
    elif os.path.isfile(item_path):
        os.remove(item_path)
list_dirs = os.listdir("../data/raw/stable_diffusion_faces/stable-diffusion-face-dataset")
print("Directories and files in 'stable_diffusion_faces':", list_dirs)
```
