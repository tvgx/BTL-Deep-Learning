```python
import os

fairfacegen_images_dir = os.path.join("..", "data", "raw", "fairfacegen", "diffusion_images")
fairfacegen_images = []

# Loop through each subdirectory in fairfacegen_images_dir
for root, dirs, files in os.walk(fairfacegen_images_dir):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            fairfacegen_images.append(os.path.join(root, file))

print(f"Total FairFaceGen images: {len(fairfacegen_images)}")
```

## Check augmentation techniques

```python
from src.process.resize_images import load_image_with_opencv
from src.process.augment_images import augment_flip, augment_rotation, augment_shift, augment_brightness, augment_contrast, augment_add_noise, random_augmentation_pipeline
```

```python
import cv2
import random

example_image_path = real_images[0]
example_image = load_image_with_opencv(example_image_path)
plt.imshow(example_image)
plt.axis('off')
plt.show()
```

```python
fliped_image = augment_flip(example_image, flip_code=random.choice([0, 1]))
plt.imshow(fliped_image)
plt.axis('off')
plt.show()
```

```python
rotated_image = augment_rotation(example_image, angle=random.uniform(-15, 15))
plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```

```python
shifted_image = augment_shift(example_image, shift_x=random.uniform(-0.1, 0.1), shift_y=random.uniform(-0.1, 0.1))
plt.imshow(shifted_image)
plt.axis('off')
plt.show()
```

```python
brightness_image = augment_brightness(example_image, factor=150)
plt.imshow(brightness_image)
plt.axis('off')
plt.show()
```

```python
constrast_image = augment_contrast(example_image, factor=1.5)
plt.imshow(constrast_image)
plt.axis('off')
plt.show()
```

```python
added_noise_image = augment_add_noise(example_image, std_dev=25)
plt.imshow(added_noise_image)
plt.axis('off')
plt.show()
```

```python
random_augmentation_pipeline_image = random_augmentation_pipeline(example_image)
plt.imshow(random_augmentation_pipeline_image)
plt.axis('off')
plt.show()
```
